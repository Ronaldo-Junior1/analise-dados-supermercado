Tendo isso em vista, vamos verificar quais produtos específicos dão mais prejuízo e quais dão mais lucro dentro dessas categorias.

```python
colunas_foco = ['Category', 'Product_Name', 'Sales', 'Quantity', 'Discount', 'Profit']
maiores_lucros_cat = df.nlargest(10, 'Profit')[colunas_foco]
maiores_prejuizos_cat = df.nsmallest(10, 'Profit')[colunas_foco]

print("Maiores Prejuízos por Categoria:")
display(maiores_prejuizos_cat)

print("\nMaiores Lucros por Categoria:")
display(maiores_lucros_cat)
```

Agora, para entender o motivo desses prejuízos gigantes (como o de -6599), geralmente a causa está associada a altos descontos (Discount). Então vamos verificar se esse é um comportamento comum, onde os maiores prejuizos são nas maiores quantidades de desconto.

```python
# Filtrando as vendas com prejuízo pior que -1000
outliers_negativos = df[df['Profit'] < -1000]

colunas_interesse = ['Product_Name', 'Segment', 'Category', 'Sales', 'Quantity', 'Discount', 'Profit']
outliers_negativos[colunas_interesse].sort_values(by='Profit').head(30)
```

Dos maiores prejuízos listados, a grande maioria foi causada por descontos de 0.7 (70%) e 0.8 (80%).

O maior prejuízo de todos (-6.599,97) ocorreu em uma venda de impressoras 3D com 70% de desconto.

Os Binding Systemsaparecem várias vezes dando prejuízos entre -1.000 e -3.700, quase todos com 80% de desconto.

O problema afeta produtos de alto valor agregado no caso tecnologia
Dar 80% de desconto em uma caneta gera um prejuízo de centavos. Mas a tabela mostra que a empresa está dando 50% a 70% de desconto em produtos caríssimos, como Impressoras 3D, Impressoras a Laser e Sistemas de Videoconferência.

A venda do sistema Cisco gerou 22.638 em receita, mas como teve 50% de desconto, resultou em um prejuízo de -1.811. Isso indica que a margem original do produto não suporta um corte de preço dessa magnitude.

Quando o desconto é alto, o cliente tende a comprar mais. Veja a mesa de conferência (Chromcraft): a empresa vendeu 13 unidades com 40% de desconto. O resultado foi um prejuízo de -1.862.

```python
# Filtrando as vendas com lucro maior que 1000
outliers_positivos = df[df['Profit'] > 1000]

outliers_positivos[colunas_interesse].sort_values(by='Profit').tail(30)
```

Dos maiores lucros listados, a esmagadora maioria foi alcançada com desconto zero (0%). O maior lucro de todos (8.399,97) ocorreu em uma venda de copiadoras avançadas com 0% de desconto e receita de 17.499.

O equipamento Canon imageCLASS 2200 Advanced Copier aparece quatro vezes dominando a lista, o que prova que os produtos de alto valor agregado no segmento de Tecnologia possuem uma margem excelente quando o preço original é mantido. As informações batem perfeitamente com a análise de prejuízos, comprovando que o problema definitivamente não está nos produtos em si, mas na política agressiva de preços.

Os mesmos Sistemas de Encadernação (Binding Systems) que afundaram o caixa com prejuízos de até -3.700 sob 80% de desconto, aparecem agora como os maiores lucros, gerando de 4.946, 4.630 e 2.504 quando vendidos a preço cheio.

A lógica do volume também se confirma, agindo agora como um multiplicador positivo: quando o cliente compra em grande quantidade sem a aplicação de descontos altos, o retorno financeiro aumenta.

O modelo GBC Ibimaster 500 exemplifica isso muito bem, com a venda de 13 unidades a 0% de desconto gerando o quarto maior lucro geral. A base mostra que a estrutura de custos da loja é sólida e os produtos são rentáveis por natureza, suportando no máximo um corte brando, já que o único desconto presente nesse top 10 foi de 20% em uma das vendas da Canon, que ainda conseguiu preservar um lucro robusto de 3.919.

##Analisando as Regiões com baixa Rentabilidade ao Longo dos Anos



###Verificando Quantos Anos tem o nosso DataSet

```python
#. Exibindo os anos que a nossa Base tem de Registo
print(f"{df["Year"].value_counts().sort_values(ascending=True)}\n")

#. Exibindo qual que é a média de dados para cada ano no DataSet
print(f"A média de dados para cada ano é de: {df["Year"].value_counts().mean()}")
```

- Podemos ver que nossa Base possui 4 anos. Desses 4 anos, 2014 e 2015 possuem uma quantidade de dados abaixo da média, o que significa que esses anos tem uma maior probabilidade de sofrer com Outliers, caso existam.

### Analisando as regiões com menos Rentabilidade ao Longo dos Anos

```python
#. Filtrando a Rentabilidade por Região dos 4 anos
rentabilidade_regiao = df.groupby('Region').Profit.mean()

#. Filtrando a Rentabilidade por Região e Ano
rentabilidade_regiao_anual = pd.crosstab(df["Region"], df["Year"], aggfunc = 'mean', values = df["Profit"])

#. Exibindo a Rentabilidade considerando os 4 anos de DataSet
print(f"Rentabilidade Geral:\n{rentabilidade_regiao.sort_values(ascending=True)}\n\n")

#. Exibindo a Rentabilidade por Região em 2014
print(f"Rentabilidade Anual 2014:\n{rentabilidade_regiao_anual[2014].sort_values(ascending=True)}\n\n")

#. Exibindo a Rentabilidade por Região em 2015
print(f"Rentabilidade Anual 2015:\n{rentabilidade_regiao_anual[2015].sort_values(ascending=True)}\n\n")

#. Exibindo a Rentabilidade por Região em 2016
print(f"Rentabilidade Anual 2016:\n{rentabilidade_regiao_anual[2016].sort_values(ascending=True)}\n\n")

#. Exibindo a Rentabilidade por Região em 2017
print(f"Rentabilidade Anual 2017:\n{rentabilidade_regiao_anual[2017].sort_values(ascending=True)}\n\n")
```

- Podemos ver que, no geral, as regiões com menos Rentabilidade são a Central e South, elas oculpam a última posição os anos de 2015 e 2017. A região Central está entre as últimas por 3 anos (2014, 2015 e 2017), então faz sentido ela estar como uma das menos rentáveis na análise geral. Porém, a região South foi a região mais rentável por dois anos (2014 e 2016) e mesmo assim, não conseguiu melhorar sua posição. Inicialmente,podemos pensar que as demais regiões conseguiram lucrar mais ao longo dos demais anos. Mas, vamos analisar mais detalhes.

### Analisando a quantidade de Dados das Regiões ao Longo dos Anos

```python
#. Descobrindo a quantidade de dados para as regiões menos rentáveis do DataSet ao longo dos anos
print(f"Quantidade de dados para a região Central do ano de 2014:\n{df[(df['Year'] == 2014) & (df['Region'] == 'Central')].shape}\n")
print(f"Quantidade de dados para a região West do ano de 2014:\n{df[(df['Year'] == 2014) & (df['Region'] == 'West')].shape}\n")

print(f"Quantidade de dados para a região South do ano de 2015:\n{df[(df['Year'] == 2015) & (df['Region'] == 'South')].shape}\n")
print(f"Quantidade de dados para a região Central do ano de 2015:\n{df[(df['Year'] == 2015) & (df['Region'] == 'Central')].shape}\n")

print(f"Quantidade de dados para a região East do ano de 2016:\n{df[(df['Year'] == 2016) & (df['Region'] == 'East')].shape}\n")
print(f"Quantidade de dados para a região West do ano de 2016:\n{df[(df['Year'] == 2016) & (df['Region'] == 'West')].shape}\n")

print(f"Quantidade de dados para a região Central do ano de 2017:\n{df[(df['Year'] == 2017) & (df['Region'] == 'Central')].shape}\n")
print(f"Quantidade de dados para a região South do ano de 2017:\n{df[(df['Year'] == 2017) & (df['Region'] == 'South')].shape}\n")
```

### Analisando os Dados Estatísticos das Regiões ao Longo dos Anos

```python
#. Exibindo os dados estatísticos da Região com a menor rentabilidade do Ano de 2014
df[(df['Year'] == 2014) & (df['Region'] == 'Central')].describe()
```

- Pela descrição dos dados podemos ver que essa região é altamente impactada por Outliers tendo em vista que o valor mínimo é muito discrepante dos demais valores, assim como o valro máximo. Já que 75% dos dados são menores que 14. Entretanto, analisando o intervalo interquartil, podemos ver que realmente essa região lucrou muito pouco. Uma das hipótes para esse baixo rendimento é quantidade de desconto que foi aplicado nas vendas. O que podemos verificar a seguir

```python
#. Exibindo a média de desconto aplicado para cada categoria
print(f"{df[(df['Year'] == 2014) & (df['Region'] == 'Central')].groupby('Category').Discount.mean()}\n")

#. Exibindo o desconto máximo para cada categoria
print(f"{df[(df['Year'] == 2014) & (df['Region'] == 'Central')].groupby('Category').Discount.max()}\n")

#. Verificando a frequência dos descontos que foram aplicados
print(f"{df[(df['Year'] == 2014) & (df['Region'] == 'Central')]['Discount'].value_counts().sort_values(ascending=False)}\n")

#. Pegando a quantidade de Vendas que foram realizadas com descontos acima de 50%
print(f"{df[(df['Year'] == 2014) & (df['Region'] == 'Central') & (df['Discount'] > 0.5)].shape[0]}\n")

#. Exibindo o lucro que foi produzido por essas vendas com o desconto maior que 50%
print(f"{df[(df['Year'] == 2014) & (df['Region'] == 'Central') & (df['Discount'] > 0.5)]['Profit'].sum()}\n")

#. Exibindo o lucro total dessa região nesse ano
print(f"{df[(df['Year'] == 2014) & (df['Region'] == 'Central')]['Profit'].sum()}\n")

#. Verificando a Rentabilidade por cada categoria
print(f"{df[(df['Year'] == 2014) & (df['Region'] == 'Central')].groupby('Category').Profit.sum()}\n")

#. Verificando quantas vendas foram realizadas em cada categoria
print(f"{df[(df['Year'] == 2014) & (df['Region'] == 'Central')]['Category'].value_counts()}\n")
```

- Podemos ver que foram aplicados descontos grandes como 60% e 80%. Além disso, se somar todos os descontos acima de 50% dá 97 vendas com descontos altos, no fim, essas 97 vendas resultaram em um prejuízo de -9143. O que parece ser o motivo da região estar com baixa rentabilidade.
